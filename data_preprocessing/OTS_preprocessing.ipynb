```python
import pandas as pd
import tidytcells as tt
import ast
import os
import sys
sys.path.insert(0, '..')
from functions import sequencefunctions as seqf
import functions.myfunctions as mf
```

```python
f2 = '../data/OTS/'
outdir = '../data/OTS/'
```

```python
cols = [
    'sequence_id_beta',
    'productive_beta',
    'v_call_beta',
    'd_call_beta',
    'j_call_beta',
    'sequence_alignment_aa_beta',
    'sequence_id_alpha',
    'productive_alpha',
    'v_call_alpha',
    'd_call_alpha',
    'j_call_alpha',
    'sequence_alignment_aa_alpha'
]
```

```python
all_dfs = []
all_meta = []
i = 1

for file in os.listdir(f2):

    if file in ['.DS_Store', 'bulk_download.sh']: continue

    # print(file)
    if i%50 == 0: print(); print(i)
    i+=1

    meta = ast.literal_eval(pd.read_csv(f2 + file, dtype='object').columns[0])
    if meta['TType'] == 'gamma-delta': continue
    if meta['TSubtype'] in ['thymocytes', 'thymocyte']: continue
    if meta['Sequences'] == 0: continue
    meta = pd.DataFrame.from_dict(meta, orient='index').T
    cols1 = ['Run', 'Author', 'Disease', 'CancerType', 'Age', 'Longitudinal',
    'Species', 'Strain', 'TSource', 'TType', 'TSubtype', 'Treatment',
    'Subject']
    all_meta.append(meta)
    
    df = pd.read_csv(f2 + file, header=1, usecols=cols)
    df = df[cols]
    df = df.loc[(df['productive_beta'] == 'T') & (df['productive_alpha'] == 'T')]
    
    df = df.dropna(subset=['v_call_beta', 'j_call_beta', 'v_call_alpha', 'j_call_alpha'], how='any')
    df = df.dropna(subset=['sequence_alignment_aa_beta', 'sequence_alignment_aa_alpha'], how='any')
    df = df.drop_duplicates(subset=['sequence_alignment_aa_beta', 'sequence_alignment_aa_alpha'])
    s1 = df.shape
    
    for c in ['v_call_beta', 'j_call_beta', 'v_call_alpha', 'j_call_alpha']:
        # print(c)
        df[c] = df.apply(
            lambda row: pd.NA if type(row[c]) != str else tt.tcr.standardise(
                gene_name=row[c],
                species='HomoSapiens',
                precision='gene'
            ),
            axis=1
        )
    df = df.dropna(subset=['v_call_beta', 'j_call_beta', 'v_call_alpha', 'j_call_alpha'], how='any')
    print('shape before tidytcells: ', s1, '; shape after tidytcells: ', df.shape)

    for c in ['sequence_alignment_aa_beta', 'sequence_alignment_aa_alpha']:
        # print(c)
        c1 = 'cdr3' + c.split('_')[-1] + '_with_gaps'
        df[c1] = df.apply(
            lambda row: pd.NA if type(row[c]) != str else seqf.get_cdr3_seq_with_gaps(row[c]),
            axis=1
        )
    s1 = df.shape
    df = df.dropna(subset=['cdr3beta_with_gaps', 'cdr3alpha_with_gaps'], how='any')
    df = df.loc[(df['cdr3beta_with_gaps'].str[0] == 'C') & (df['cdr3beta_with_gaps'].str[0] == 'C')]
    df['len_cdr3a'] = [len(x) for x in df['cdr3alpha_with_gaps']]
    df['len_cdr3b'] = [len(x) for x in df['cdr3beta_with_gaps']]
    df = df.loc[(df.len_cdr3a < 20) & (df.len_cdr3b < 20)]
    df = mf.prepare_data(df, col1 = 'cdr3alpha_with_gaps', col2='cdr3beta_with_gaps', type = 'cdr3', max_len1=19, max_len2=19)
    
    print('shape before cdr3: ', s1, '; shape after cdr3: ', df.shape)

    df[cols1] = list(meta[cols1].values)*df.shape[0]
    all_dfs.append(df)
```

```python
all_meta = pd.concat(all_meta)
```

```python
for c in all_meta.columns:
    print(c)
    print(all_meta[c].value_counts())
```

```python
big_df = pd.concat(all_dfs)
```

```python
big_df.to_csv(outdir + 'OTS_all_studies_combined.csv.gz')
```
