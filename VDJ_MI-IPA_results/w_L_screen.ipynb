```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
os.chdir('..')
import functions.helperfunctions as hf
```

```python
ff1 = 'data/output/pairing_MI-IPA/w_L_screen_GLC/'
ff2 = 'data/output/pairing_MI-IPA/w_L_screen_YLQ/'
```

```python
allresults_GLC = []
for f in os.listdir(ff1):
    if '.csv' in f:
        print(f)
        f1 = f.replace('small_ind', 'smallInd')
        f1 = f1.replace('prop_test', 'propTest')
        properties = {x.split('-')[0]:x.split('-')[1] for x in f1.split('_') if '.csv' not in x}
        results, N = hf.load_df_TPR(ff1 + f)
        results = pd.DataFrame(np.array(np.array([x for x in results])))
        its = results.shape[1]-1
        results['test_size'] = N
        results['final_test'] = results[its]
        results['final_test_prop'] = results[its] / N
        results['L'] = properties['L']
        results['w'] = properties['weights']
        results['Epitope'] = properties['epitope']
        allresults_GLC.append(results)
```

```python
all_results_GLC = pd.concat(allresults_GLC)
```

```python
Rmelt_GLC = pd.melt(all_results_GLC, id_vars = ['L', 'w', 'final_test_prop', 'final_test', 'test_size', 'Epitope'])
```

```python
print(Rmelt_GLC)
```

```python
allresults_YLQ = []
for f in os.listdir(ff2):
    if '.csv' in f:
        print(f)
        f1 = f.replace('small_ind', 'smallInd')
        f1 = f1.replace('prop_test', 'propTest')
        properties = {x.split('-')[0]:x.split('-')[1] for x in f1.split('_') if '.csv' not in x}
        results, N = hf.load_df_TPR(ff2 + f)
        results = pd.DataFrame(np.array(np.array([x for x in results])))
        its = results.shape[1]-1
        results['test_size'] = N
        results['final_test'] = results[its]
        results['final_test_prop'] = results[its] / N
        results['L'] = properties['L']
        results['w'] = properties['weights']
        results['Epitope'] = properties['epitope']
        allresults_YLQ.append(results)
```

```python
all_results_YLQ = pd.concat(allresults_YLQ)
Rmelt_YLQ = pd.melt(all_results_YLQ, id_vars = ['L', 'w', 'final_test_prop', 'final_test', 'test_size', 'Epitope'])
```

Lineplots for everything

```python
Rmelt_GLC['value'] = Rmelt_GLC['value']/Rmelt_GLC['test_size']
Rmelt_GLC = Rmelt_GLC.sort_values(by = ['w','L'])
d = {'color':sns.color_palette('flare',5) + ['black'], 'ls' : ['-', '--', '-.', ':', '--', '-']}
g = sns.FacetGrid(Rmelt_GLC[['w','L','variable','value']], col = 'w', hue='L', hue_kws=d, col_wrap=6, legend_out=False)
g = g.map(sns.lineplot, 'variable', 'value', errorbar = 'sd').add_legend()
lgd=sns.move_legend(g, 'upper center',
    bbox_to_anchor=(.5, -0.01), ncol=6, title=r'$\lambda$', frameon=False, fontsize=14, title_fontsize=14)
for ax in g.axes:
    ax.set_xlabel('iteration', fontdict={'size':14})
    ax.set_ylabel('Precision', fontdict={'size':14})
    ax.set_title(r'$\theta = $' + ax.get_title().split('=')[1], fontsize=14)
g.fig.suptitle(Rmelt_GLC['Epitope'].unique()[0], fontsize=16)
plt.tight_layout()
plt.show()
```

```python
Rmelt_YLQ['value'] = Rmelt_YLQ['value']/Rmelt_YLQ['test_size']
Rmelt_YLQ = Rmelt_YLQ.sort_values(by = ['w','L'])
d = {'color':sns.color_palette('flare',5) + ['black'], 'ls' : ['-', '--', '-.', ':', '--', '-']}
g = sns.FacetGrid(Rmelt_YLQ[['w','L','variable','value']], col = 'w', hue='L', hue_kws=d, col_wrap=6, legend_out=False)
g = g.map(sns.lineplot, 'variable', 'value', errorbar = 'sd').add_legend()
lgd=sns.move_legend(g, 'upper center',
    bbox_to_anchor=(.5, -0.01), ncol=6, title=r'$\lambda$', frameon=False, fontsize=14, title_fontsize=14)
for ax in g.axes:
    ax.set_xlabel('iteration', fontdict={'size':14})
    ax.set_ylabel('Precision', fontdict={'size':14})
    ax.set_title(r'$\theta = $' + ax.get_title().split('=')[1], fontsize=16)
g.fig.suptitle(Rmelt_YLQ['Epitope'].unique()[0], fontsize=16)
plt.tight_layout()
plt.show()
```

A single plot to show summary performance

```python
rmelt1 = Rmelt_GLC[['final_test_prop','L','w']].drop_duplicates()
rmelt2 = Rmelt_YLQ[['final_test_prop','L','w']].drop_duplicates()

f, ax = plt.subplots(ncols=2, sharey=True, figsize=(10,4))
sns.lineplot(rmelt1, x = 'L', y = 'final_test_prop', hue = 'w', style='w', markers=True, palette = 'Oranges', 
                markersize = 10, errorbar = 'sd', ax = ax[0])
                #err_style='bars', err_kws = {'capsize':5})
sns.lineplot(rmelt2, x = 'L', y = 'final_test_prop', hue = 'w', style='w', markers=True, palette = 'Oranges', 
                markersize = 10, errorbar = 'sd', ax = ax[1])
handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend().remove()
ax[1].legend().remove()
f.legend(handles, labels, bbox_to_anchor = [0.5,0], loc='upper center', title=r'$\theta$', ncols=6, fontsize=14, title_fontsize=14)
ax[0].set_xlabel(r'$\lambda$', fontsize=14)
ax[1].set_xlabel(r'$\lambda$', fontsize=14)
ax[0].set_ylabel('Precision', fontsize=14)
ax[0].set_title(Rmelt_GLC['Epitope'].unique()[0], fontsize=14)
ax[1].set_title(Rmelt_YLQ['Epitope'].unique()[0], fontsize=14)
plt.tight_layout()
plt.savefig('figures/wL_summary_TPR.png',
                    bbox_inches='tight')
plt.show()
```
