```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys

sys.path.insert(0, '..')
from PairingVDJdb_MI import _define_individuals
import functions.myfunctions as mf
import tidytcells as tt
```

```python
vdj = pd.read_csv('../data/vdj_cleaned_subset_for_MI.csv', index_col = 0)
vdj = vdj.loc[vdj['Epitope'] != 'KLGGALQAK'] # because too big - takes forever
vdj = mf.prepare_data(vdj, col1 = 'cdr3a_IMGTgaps', col2='cdr3b_IMGTgaps', type = 'cdr3').drop_duplicates(subset = ['cdr3a_IMGTgaps_padded', 'cdr3b_IMGTgaps_padded', 'Epitope'])
print(vdj.shape)

for c in ['V-a', 'V-b', 'J-a', 'J-b']:
    print(c)
    vdj[c] = vdj.apply(
        lambda row: pd.NA if type(row[c]) != str else tt.tcr.standardise(
            gene_name=row[c],
            species=row['Species'],
            precision='gene'
        ),
        axis=1
    )
vdj00 = vdj.dropna(subset=['V-a', 'V-b', 'J-a', 'J-b'])
print('shape before tidytcells: ', vdj.shape, '; shape after tidytcells: ', vdj00.shape)

vdj = vdj00.copy()
```

```python
vdj = vdj.reset_index(drop=True).reset_index(names = ['original_pair_index'])
```

```python
vdj
```

```python
known_binders = []

for ep in vdj['Epitope'].unique():
   known_binders.append(_define_individuals(vdj, ep, ind = 'no'))

known_binders
```

```python
known_binders = pd.concat(known_binders)
assert known_binders.shape[0] == vdj.shape[0]
```

```python
known_binders = known_binders[['original_pair_index', 'CDR3-a', 'CDR3-b', 'Epitope', 'MHC A', 'subject-PMID']]
```

```python
known_binders['binder'] = 1
```

```python
known_binders.head()
```

```python
shuffled = known_binders.copy()
```

```python
shuffled1 = []

for ep in shuffled['Epitope'].unique():
    s_ep = shuffled.loc[shuffled['Epitope'] == ep]
    for subj in s_ep['subject-PMID'].unique():
        s_ep_subj = s_ep.loc[s_ep['subject-PMID'] == subj]
        betas = s_ep_subj['CDR3-b'].unique().tolist()
        for a in s_ep_subj['CDR3-a'].unique():
            s_ep_subj_a = s_ep_subj.loc[s_ep_subj['CDR3-a'] == a]
            for b in betas:
                s_ep_subj_a1 = s_ep_subj_a.copy()
                s_ep_subj_a1['CDR3-b_shuffle'] = b
                shuffled1.append(s_ep_subj_a1)
        
```

```python
shuffled1 = pd.concat(shuffled1)
shuffled1 = shuffled1.drop_duplicates()
shuffled1
```

```python
shuffled1['pair_name'] = shuffled1['CDR3-a'] + '::' + shuffled1['CDR3-b_shuffle'] + '::' + shuffled1['subject-PMID']
shuffled1
```

```python
known_binders['pair_name'] = known_binders['CDR3-a'] + '::' + known_binders['CDR3-b'] + '::' + known_binders['subject-PMID']
```

```python
shuffled1 = shuffled1.loc[~shuffled1['pair_name'].isin(known_binders['pair_name'])]
shuffled1
```

```python
shuffled1['binder'] = 0
shuffled1['CDR3-b'] = shuffled1['CDR3-b_shuffle']
shuffled1 = shuffled1.drop('CDR3-b_shuffle', axis=1)
shuffled1
```

```python
all_pairs = pd.concat([known_binders, shuffled1])
all_pairs.to_csv('../data/output/benchmarking/benchmarking_pairs.csv.gz')
```
