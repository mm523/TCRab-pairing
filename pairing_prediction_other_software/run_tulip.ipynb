```python
df = pd.read_csv('../TCRab-pairing/data/output/benchmarking/benchmarking_pairs.csv.gz', index_col = 0)
df = df.rename(columns = {'Epitope':'peptide', 'CDR3-a':'CDR3a', 'CDR3-b':'CDR3b', 'MHC A':'MHC'})
```

```python
df['MHC'].value_counts()
```

```python
df['MHC'] = [x.split(':')[0] for x in df['MHC']]
df['MHC'].value_counts()
```

```python
print(df.shape)
df = df[['CDR3a', 'CDR3b', 'peptide', 'MHC', 'binder']].drop_duplicates(subset=['CDR3a', 'CDR3b', 'peptide', 'MHC'], 
                                                                        keep = 'first')
print(df.shape)
assert df[['CDR3a', 'CDR3b', 'peptide', 'MHC']].value_counts().max() == 1
assert df[['CDR3a', 'CDR3b', 'peptide']].value_counts().max() == 1
df.to_csv('../AB-interactions/data/output/benchmarking/TULIP_no-small-study/benchmarking_pairs_TULIP_no-small-study.csv')
```
