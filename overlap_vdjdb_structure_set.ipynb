```python
vdjdb_tcrs = list(zip(vdjdb['CDR3-a'], vdjdb['CDR3-b']))
vdjdb_complexes = list(zip(vdjdb['CDR3-a'], vdjdb['CDR3-b'], vdjdb['Epitope']))
```

```python
tcrdb['tcr_in_vdj_set'] = [
    (a.replace('-',''), b.replace('-','')) in vdjdb_tcrs for a,b in zip(tcrdb['cdr3a'], tcrdb['cdr3b'])
]
tcrdb['complex_in_vdj_set'] = [
    (a.replace('-',''), b.replace('-',''), e) in vdjdb_tcrs for a,b,e in zip(tcrdb['cdr3a'], tcrdb['cdr3b'], tcrdb['epitope_aa'])
]
```
